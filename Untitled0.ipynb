```python
data_transform = df.copy()
ordinal_encoder = OrdinalEncoder()
data_transform[catcolumns] = ordinal_encoder.fit_transform(data_transform[catcolumns])
preclass = {"NObeyesdad": {"Insufficient_Weight": 0,
                          "Normal_Weight": 1,
                         'Overweight_Level_I': 2,
                          'Overweight_Level_II': 3,
                          'Obesity_Type_I': 4,
                          'Obesity_Type_II': 5,
                          'Obesity_Type_III': 6,}
                         }
data_transform.replace(preclass, inplace=True)
data_transform
```

```python
data_transform.dtypes
```

```python
data_transform.describe()
```

```python
# Split the dataset into features (X) and target (y)
X = data_transform.drop('NObeyesdad', axis=1)
y = data_transform['NObeyesdad']
print(X.shape)
print(y.shape)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# standard scale data
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)

# tested MinMaxScaler as KNN historically does better with MinMax
#mm = MinMaxScaler()
#X_train_mm_scaled = ss.fit_transform(X_train)
#X_test_mm_scaled = ss.transform(X_test)
```

```python
#CLASSIFICATION
```

```python
#1. Decision Tree Classifier
dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train_scaled, y_train)
dt_predictions = dt_clf.predict(X_test_scaled)
print("Decision Tree Classifier:")
print(classification_report(y_test, dt_predictions))
print(confusion_matrix(y_test,dt_predictions, labels = [1,0]))
c1=confusion_matrix(y_test,dt_predictions)
sn.heatmap(c1, annot=True)
p.show()
#confusion_matrix(y_test, prediction, labels = [1,0])
```

```python
#2. Random Forest Classification
rf_clf = RandomForestClassifier()
rf_clf.fit(X_train_scaled, y_train)
rf_predictions = rf_clf.predict(X_test_scaled)
print("Random Forest Classifier:")
print(classification_report(y_test, rf_predictions))
print(confusion_matrix(y_test,rf_predictions, labels = [1,0]))
c2=confusion_matrix(y_test,rf_predictions)
sn.heatmap(c2, annot=True)
p.show()
#confusion_matrix(y_test, prediction, labels = [1,0])
```

```python
#3. Support Vector Machine (SVM) Classifier
svm_clf = SVC()
svm_clf.fit(X_train_scaled, y_train)
svm_predictions = svm_clf.predict(X_test_scaled)
print("  Support Vector Machine (SVM)Classifier:")
print(classification_report(y_test, svm_predictions))
print(confusion_matrix(y_test,svm_predictions, labels = [1,0]))
c3=confusion_matrix(y_test,svm_predictions)
sn.heatmap(c3, annot=True)
p.show()
#confusion_matrix(y_test, prediction, labels = [1,0])
```

```python
#4.'KNN', KNeighborsClassification
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train_scaled, y_train)
knn_predictions = knn_clf.predict(X_test_scaled)
print(" KNeighbors(KNN)Classifier:")
print(classification_report(y_test, knn_predictions))
print(confusion_matrix(y_test,knn_predictions, labels = [1,0]))
c4=confusion_matrix(y_test,knn_predictions)
sn.heatmap(c4, annot=True)
p.show()
#confusion_matrix(y_test, prediction, labels = [1,0])
```

```python
#5.GaussianNaive_bayes Classification
from sklearn.naive_bayes import GaussianNB
nb_clf = GaussianNB()
nb_clf.fit(X_train_scaled, y_train)
nb_predictions = nb_clf.predict(X_test_scaled)
print(" GaussianNaive_bayes Classifier:")
print(classification_report(y_test, nb_predictions))
print(confusion_matrix(y_test,nb_predictions, labels = [1,0]))
c5=confusion_matrix(y_test,nb_predictions)
sn.heatmap(c5, annot=True)
p.show()
#confusion_matrix(y_test, prediction, labels = [1,0])
```

```python
# 1. Logistic Regression Classifier
lr_clf = LogisticRegression()
lr_clf.fit(X_train_scaled, y_train)
lr_predictions = lr_clf.predict(X_test_scaled)
print("Logistic Regression Classifier:")
print(classification_report(y_test, lr_predictions))
print(confusion_matrix(y_test,lr_predictions, labels = [1,0]))
c6=confusion_matrix(y_test,lr_predictions)
sn.heatmap(c6, annot=True)
p.show()
#confusion_matrix(y_test, prediction, labels = [1,0])
```

```python
#COMPARISON BETWEEN DIFFERENT MODELS RESULT USING BOXPLOT
# Application of all Machine Learning methods
models = []
models.append(('LR', LogisticRegression()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
models.append(('RF',RandomForestClassifier()))
```

```python
# evaluate each model in turn
from sklearn import model_selection
from sklearn.model_selection import StratifiedKFold

results = []
names = []
scoring = 'accuracy'
for name, model in models:
	kfold = model_selection.KFold(n_splits=10, random_state=None,shuffle=False )
	cv_results = model_selection.cross_val_score(model, X_train_scaled, y_train, cv=kfold, scoring=scoring)
	results.append(cv_results)
	names.append(name)
	msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
	print(msg)
# boxplot algorithm comparison
fig = plt.figure()
fig.suptitle('Comparison between different MLAs')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
